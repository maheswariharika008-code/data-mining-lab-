```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity
```

```python
data = {
    "order_id": [1, 2, 3, 4],
    "customer": ["Alice", "Bob", "Alice", "David"],
    "amount": [100.0, 200.0, 150.0, 300.0],
    "tax": [10.0, 20.0, 15.0, 30.0],
    "total_amount": [110.0, 220.0, 165.0, 330.0]
}

df = pd.DataFrame(data)
print(df)
```

```python
df.describe()
```

```python
features = df[["order_id", "amount", "tax", "total_amount"]]
```

```python
euclidean_distances(features)
```

```python
cosine_similarity(features)
```
